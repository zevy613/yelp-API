```python
import json
from yelpapi import YelpAPI

with open('/Users/default.DESKTOP-0LDO0LD/.secret/yelp_api.json') as f: #change the path to match YOUR path!!
    login = json.load(f)

yelp_api = YelpAPI(login['api-key'], timeout_s=5.0)

# Page 1
results = yelp_api.search_query(location='Chicago, IL', term='Piazza')
print(type(results))
results.keys()
```

```python
# what is in the businesses key?
print(type(results['businesses']))

# how long is businesses?
print("Number of businesses ", len(results['businesses']))
```

```python
# how many entries do we have.
print("Number of results ", results['total'])
```

```python
import pandas as pd
df_0_20 = pd.DataFrame(results['businesses'])
df_0_20.head()
```

```python
# Page 2
results = yelp_api.search_query(location='Chicago, IL', term='Piazza', offset=20)
```

```python
df_20_40 = pd.DataFrame(results['businesses'])
df_20_40.head()
```

```python
## concatenate the previous results and new results. 
df_0_40 = pd.concat([df_0_20, df_20_40], ignore_index=True)
display(df_0_40.head(3), df_0_40.tail(3))
```

```python
# Save the businesses as a records-oriented JSON file
w = df_0_40.to_json(orient='records')
type(w)
w
```
