```python
## Standard Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
## Importing the OS and JSON Modules
import os,json
```

```python
df_info = pd.read_csv('Data/superhero_info - superhero_info.csv')
df_powers = pd.read_csv("Data/superhero_powers - superhero_powers.csv")

display(df_info.head())
display(df_powers.head())
```

```python
#seperate the string column 'Hero|Publisher'
df_info[['Hero', 'Publisher']] = df_info['Hero|Publisher'].str.split('|', expand=True)
df_info.drop(columns=['Hero|Publisher'], inplace=True)
df_info.head(3)
```

```python
#Lets change the Measurements column to a dictionary
df_info['Measurements'] = df_info['Measurements'].str.replace("'",'"')
df_info['Measurements'] = df_info['Measurements'].apply(json.loads)

# Check
temp = df_info.loc[0, 'Measurements']
display(type(temp), temp)
```

```python
# Lets unpack the dictionary that we just created into different columns
temp = df_info['Measurements'].apply(pd.Series)

# lets add these new columns to our data frame
df_info = pd.concat((df_info, temp), axis=1)
df_info.drop(columns = ['Measurements'], inplace=True)

# Check
df_info.head()
```

```python
# Convert the Height and Weight columns to float columns

# First lets convert the type of the column to strings
df_info['Height'] = df_info['Height'].astype('string')
df_info['Weight'] = df_info['Weight'].astype('string')

# Next we will extract the float from the string
df_info['Height'] = df_info['Height'].apply(lambda x: x.split(' ')[0])
df_info['Weight'] = df_info['Weight'].apply(lambda x: x.split(' ')[0])

# Next we will convert the columns from string to float
df_info['Height'] = df_info['Height'].astype('float')
df_info['Weight'] = df_info['Weight'].astype('float')

# Check
display(df_info.info(), df_info.head())
```

```python
# Lets rename the height and weight columns so we dont lose the units
df_info.rename(columns={'Height': 'Height_cm', 'Weight': 'Weight_kg'}, inplace=True)
df_info.head()
```

```python
df_powers['Powers_split'] = df_powers['Powers'].str.split(',', expand=False)

# exploding the column of lists
exploded = df_powers.explode('Powers_split')

## saving the unique values from the exploded column
cols_to_make = exploded['Powers_split'].dropna().unique()
```

```python
# Add a column for each power and assign a true/false value if that super hero has that power
for col in cols_to_make:    
    df_powers[col] = df_powers['Powers'].str.contains(col) # Powers

# Check
df_powers.head()
```

```python
# Lets combine the tables to make on big table
merged = pd.merge(df_info, df_powers, left_on='Hero', right_on='hero_names', how='inner')
merged.head(3)
```

## Part 2

```python
# Compare the average weight of super powers who have Super Speed to those who do not.
super_speed = merged['Super Speed'] == True

df_speed = merged[super_speed]
df_nospeed = merged[~super_speed]

print("The average weight for super heros who have super speed : ", df_speed['Weight_kg'].mean())
print("The average weight for super heros who don't have super speed : ", df_nospeed['Weight_kg'].mean())
```
